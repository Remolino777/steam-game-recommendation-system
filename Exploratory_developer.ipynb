```python
import pandas as pd
import json
import ast
```

### Cargo las base de datos


```python
data = []
with open(r'Dataset\\output_steam_games.json', 'r', encoding='utf-8') as file:
    for line in file:
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError:
            continue  # Skip the line if it can't be parsed
```

```python
df = pd.DataFrame(data)
df
```

```python
df = df.dropna()
df
```

```python
df1 = df.copy()
```

```python
# Creo el dataframe con las columnas que nesecito para poder crear las funciones.
df2 = df1[['developer','id', 'release_date', 'price']].copy()
df2
```

### Normalizo la columna price... Free to play y free to use lo volvemos a valor 0.

```python
df2['price'] = df2['price'].replace('Free to Play', 0)
```

```python
df2['price'] = df2['price'].replace('Free to Use', 0)
```
