```python
import pandas as pd
import json
import ast
```

```python
#Cargamos los datos.
data = []
with open(r'Dataset\australian_user_reviews.json', 'r', encoding='utf-8') as file:
    for line in file:
        data.append(eval(line))
```

```python
df = pd.DataFrame(data)
df.info()
```

```python
df.describe()
```

```python
# Elimino la columna 'user_url'
df = df.drop(['user_url'], axis=1)
df.head(5)
```

### Exploro la columna reviews



```python
# df_items = pd.json_normalize(df['items'].explode())

df_r = pd.json_normalize(df['reviews'].explode())
df_r
```

```python
#obtengo un df con las columnas ['item_id','review'] con despues lo salvo en un archivo csv
df_r2 = df_r[['item_id','review']]
df_r2.head(3)
```

```python
df_r2.to_csv(r'Dataset\Procesado\user_reviews.csv', index=False)
```
